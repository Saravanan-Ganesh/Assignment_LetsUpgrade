```python
"""Questions 2:

How to create a series from a numpy array?"""
                                                                                                        # Answer for Question 2
# Solution

import pandas as pd 
import numpy as np 
  
array = np.array(['kotagiri', 'ooty', 'coimbatore', 'chennai', 'coonoor','delhi'])                  # array created by numpy 
  
series = pd.Series(array)                                                                           # series created by pandas 
print(series)
```
